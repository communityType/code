```python
def neighbors_dict(filenames):
    '''
        use twitter .jsonl file to generate a {node:neighbors} dict
    '''
    nei = {}
    id_list = []
    for fname in filenames:
        with open(fname, 'r') as f: 
            for line in f: 
                tweet = json.loads(line) 
                if tweet.get('id') not in id_list:
                    retweet = tweet.get('retweeted_status')
                    id_list.append(tweet.get('id'))
                    if retweet:
                        source = retweet.get('user').get('id_str')
                        target = tweet.get('user').get('id_str')

                        if target in nei.keys():
                            nei[target].append(source)
                        else:
                            nei[target] = [source]
    return nei;
```

```python
def list_to_set(G,par,network_type = 'nx'):
    result = []
    if network_type == 'nx':
        node_order = [v for v in G.nodes()]
    elif network_type == 'gt': 
        node_order = [G.vp.name[n] for n in G.vertices()]
    
    for item in set(par):
        result.append({node_order[i] for i in range(len(par)) if par[i] == item})
        
    return result;
```

```python
def sort_par(par):
    '''
        sort the given partition based on group size
        return dict {par:sorted_par}
    '''
    count = Counter(par)
    sp = [k for k, v in sorted(count.items(), key=lambda item: item[1],reverse = True)]
    result = {sp[i]:i for i in range(len(sp))}
    return result;
```

```python
def get_weighted_matrix(g,state):
    B = state.get_B()
    m = np.zeros((B,B))
    ecount = g.ep['count']
    par = state.get_blocks()
    spar = sort_par(par)
    for e in g.edges():
        s = e.source()
        t = e.target()
        
        m[spar[par[s]]][spar[par[t]]] += ecount[e]
        
    
    #remove all 0 entries
    if state.get_B() != state.get_nonempty_B():
        idx = np.where(m[0]==0)[0][0]
        
        m = np.delete(m,idx,axis = 1)
        m = np.delete(m,idx,axis = 0)
    
    plt.matshow(m)
    return m;
```

## 2.2 Density Matrix

```python
def get_weighted_density(g,state):
    B = state.get_B()
    m = np.zeros((B,B))
    ecount = g.ep['count']
    par = state.get_blocks()
    count_par = Counter(par)
    spar = sort_par(par)
    for e in g.edges():
        s = e.source()
        t = e.target()
        
        m[spar[par[s]]][spar[par[t]]] += ecount[e]/count_par[par[s]]/count_par[par[t]]
        
    
    #remove all 0 entries
    if state.get_B() != state.get_nonempty_B():
        idx = np.where(m[0]==0)[0][0]
        
        m = np.delete(m,idx,axis = 1)
        m = np.delete(m,idx,axis = 0)
    
    plt.matshow(m,)
    return m;
```

```python
def avg_degree(g,par_name,typ = "total"):
    avg = []
    for i in set(g.vp[par_name].get_array()):
        vlist = [n for n in g.get_vertices() if g.vp[par_name][n] == i]
        if typ == "total":
            deg = G_gt.get_total_degrees(vlist,eweight = G_gt.ep['count']) 
        elif typ == "in":
            deg = G_gt.get_in_degrees(vlist,eweight = G_gt.ep['count'])
        elif typ == "out":
            deg = G_gt.get_out_degrees(vlist,eweight = G_gt.ep['count'])
        avg.append(np.mean(deg))
    return avg;
```

## 3.1 Module interation modifs

We can characterize the interaction between communities based on their pairwise community densities $\omega_{rs} = \frac{1}{N_rN_s}\sum_{i\in s, j \in r} A_{ij}$ for community $r$ and $s$.

Then we can classify the community interactions as one of three types:
$M_{rs}=\begin{cases}
 &M_{\text{assortative}} \text{ if } \min (\omega_{rr},\omega_{ss}) > \omega_{rs} \\
 &M_{\text{core_periphery}} \text{ if } \max (\omega_{rr},\omega_{ss}) > \omega_{rs} > \min (\omega_{rr},\omega_{ss})\\
 &M_{\text{disassortative}} \text{ if } \omega_{rs} > \max (\omega_{rr},\omega_{ss})
\end{cases}$

Given a k-community partition, each community $r$ participates in $k-1$ interactions $M_r = {m_{1},...,m_{k-1}}$.
We can calculate $P_{rt}$ as the fraction of all $k-1$ interations that are of type $t$.


```python
def interaction_type(wrr,wss,wrs):
    if min(wrr,wss) >= wrs:
        return "Assortative";
    elif wrs >= max(wrr,wss):
        return "Disassortative";
    else:
        return "Core-Periphery";
    
def interaction_types(den):
    '''
        calculate Mrs
    '''
    N,_ = den.shape
    M = np.ndarray(shape=(N,N),dtype = object)
    
    for i in range(N):
        for j in range(N):
            if i != j:
                M[i,j] = interaction_type(den[i,i],den[j,j],den[i,j])
    return M;

def interaction_type_int(wrr,wss,wrs):
    if min(wrr,wss) >= wrs:
        #Assortative
        return 0.0;
    elif wrs >= max(wrr,wss):
        #Disassortative
        return 1.0;
    else:
        #Core-Periphery
        return 0.5;
    
def interaction_types_int(den):
    '''
        calculate Mrs
    '''
    N,_ = den.shape
    M = np.ndarray(shape=(N,N))
    
    for i in range(N):
        for j in range(N):
            if i != j:
                M[i,j] = interaction_type_int(den[i,i],den[j,j],den[i,j])
            else:
                M[i,j] = -1.0
    plt.figure(figsize=(8, 6))            
    plt.matshow(M, fignum=1)
    plt.colorbar()
    plt.title("Interaction Type Matrix")
    return;

def count_rt(count,typ):
    if typ in count.keys():
        return count[typ];
    else:
        return 0;

def interaction_prob(M_rs):
    '''
        calculate Prt
    '''
    N,_ = M_rs.shape
    Pout = np.ndarray(shape=(N,3))
    Pin = np.ndarray(shape=(N,3))
    
    for i in range(N):
        #out behaviour,i.e. from i to others
        count = Counter(np.delete(M_rs[i,:], [i]))
        Pout[i,0] = count_rt(count,"Assortative")/(N-1)
        Pout[i,1] = count_rt(count,"Core-Periphery")/(N-1)
        Pout[i,2] = count_rt(count,"Disassortative")/(N-1)
    
        #in behaviour,i.e. from others to i
        count = Counter(np.delete(M_rs[:,i], [i]))
        Pin[i,0] = count_rt(count,"Assortative")/(N-1)
        Pin[i,1] = count_rt(count,"Core-Periphery")/(N-1)
        Pin[i,2] = count_rt(count,"Disassortative")/(N-1)
    
    return Pout,Pin;

def interaction_type_prob(M_rs):
    '''
        calculate Pt over all interactions
    '''
    N,_ = M_rs.shape
    Pt = np.ndarray(shape=(1,3))
    for i in range(N):
        count = Counter(np.delete(M_rs[i,:], [i]))
        Pt[0,0] += count_rt(count,"Assortative")
        Pt[0,1] += count_rt(count,"Core-Periphery")
        Pt[0,2] += count_rt(count,"Disassortative")
        
    Pt[0,0] = Pt[0,0]/(N-1)/(N-1)
    Pt[0,1] = Pt[0,1]/(N-1)/(N-1)
    Pt[0,2] = Pt[0,2]/(N-1)/(N-1)
    return Pt;

def threshold_count(arr,threshold):
    return np.mean(arr>=threshold);
```

```python
def statistics_Network(den,graph,method):
    print("---------------------------------------------------------------------")
    print("Statistics of network„Äê",graph,"„Äëcommunity found by„Äê",method,"„Äë")
    print("---------------------------------------------------------------------")
    interaction_types_int(den)
    
    Mrs = interaction_types(den)
    Pt = interaction_type_prob(Mrs)
    print("There are %.2f interactions are Assortative"%Pt[0,0])
    print("There are %.2f interactions are Core-Periphery"%Pt[0,1])
    print("There are %.2f interactions are Disassortative"%Pt[0,2])
    
    Prt_out, Prt_in = interaction_prob(Mrs)
    print("---------------------------------------------------------------------")
    print("Among all communities, %.2f has all assortative out-interactions"%threshold_count(Prt_out[:,0],1))
    print("Among all communities, %.2f has 0.5 assortative out-iinteractions"%threshold_count(Prt_out[:,0],0.5))
    print("---------------------------------------------------------------------")
    print("Among all communities, %.2f has all assortative in-interactions"%threshold_count(Prt_in[:,0],1))
    print("Among all communities, %.2f has 0.5 assortative in-interactions"%threshold_count(Prt_in[:,0],0.5))
    
    w = 16
    h = 6
    d = 70
    figure, (ax1, ax2) = plt.subplots(1, 2,figsize=(w, h))

    ax1.plot(range(len(Prt_out[:,0])),Prt_out[:,0],"o")
    ax1.plot(range(len(Prt_out[:,1])),Prt_out[:,1],"+")
    ax1.plot(range(len(Prt_out[:,2])),Prt_out[:,2],"*")
    ax1.set_title('P-out')


    ax2.plot(Prt_in[:,0],"o",label = "Assortative")
    ax2.plot(Prt_in[:,1],"+",label = "Core-Periphery")
    ax2.plot(Prt_in[:,2],"*",label = "Disassortative")
    ax2.set_title('P-in')
    ax2.legend(loc = "upper right", bbox_to_anchor=(1.3, 1))
    plt.show()
    return;
    
```

## 1. advogato

A network of trust relationships among users on Advogato, an online community of open source software developers. Edge direction indicates that node i trusts node j, and edge weight denotes one of four increasing levels of declared trust from i to j: observer (0.4), apprentice (0.6), journeyer (0.8), and master (1.0).

```python
def convert_gt_to_nx(G,directed = False, counts = True, vp_name = None, ep_name = None):
    if directed:
        g = nx.DiGraph()
    else:
        g = nx.Graph()
        
    for e in G.edges():
        s,t = e
        if counts:
            g.add_edge(G.vp[vp_name][s],G.vp[vp_name][t],weight = G.ep[ep_name][e])
        else:
            g.add_edge(G.vp[vp_name][s],G.vp[vp_name][t])
        
    return g;
```
