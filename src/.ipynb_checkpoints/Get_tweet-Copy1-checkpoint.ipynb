```python
from queue import Queue
from threading import Thread

class Listener(tweepy.StreamListener):

    def __init__(self, q = Queue()):
        super().__init__()
        self.q = q
        for i in range(4):
            t = Thread(target=self.do_stuff)
            t.daemon = True
            t.start()

#     def on_status(self, status):
#         <my code here>

    def do_stuff(self):
        while True:
            self.q.get()
            self.q.task_done()
```

```python
def neighbors_dict(filenames):
    '''
        use twitter .jsonl file to generate a {node:neighbors} dict
    '''
    nei = {}
    id_list = []
    for fname in filenames:
        with open(fname, 'r') as f: 
            for line in f: 
                tweet = json.loads(line) 
                if tweet.get('id') not in id_list:
                    retweet = tweet.get('retweeted_status')
                    id_list.append(tweet.get('id'))
                    if retweet:
                        source = retweet.get('user').get('id_str')
                        target = tweet.get('user').get('id_str')

                        if target in nei.keys():
                            nei[target].append(source)
                        else:
                            nei[target] = [source]
    return nei;
```

## Assortativity

```python
def list_to_set(G,par,network_type = 'nx'):
    result = []
    if network_type == 'nx':
        node_order = [v for v in G.nodes()]
    elif network_type == 'gt': 
        node_order = [G.vp.name[n] for n in G.vertices()]
    
    for item in set(par):
        result.append({node_order[i] for i in range(len(par)) if par[i] == item})
        
    return result;
```

## weighted matrix

```python
def get_weighted_matrix(G,par,network_type = 'nx'):
    B = max(par)+1
    m = np.zeros((B,B))
    
    if network_type == 'nx':
        node_order = [v for v in G.nodes()]
    elif network_type == 'gt': 
        node_order = [n for n in G.vertices()]
        
    for e in G.edges():
        u,v = e
        iu = node_order.index(u)
        iv = node_order.index(v)
        if network_type == 'nx':
            m[par[iu]][par[iv]] += G[u][v]['weight']
        elif network_type == 'gt': 
            m[par[iu]][par[iv]] += G.ep['count'][e]
            
    
    return m.astype(int);
```

```python
def convert_gt_to_nx(G,directed = False, counts = True):
    if directed:
        g = nx.DiGraph()
    else:
        g = nx.Graph()
        
    for e in G.edges():
        s,t = e
        g.add_edge(G.vp['name'][s],G.vp['name'][t],weight = G.ep['count'][e])
        
    return g;
```
