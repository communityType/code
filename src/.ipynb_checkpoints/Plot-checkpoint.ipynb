## Fig 1: Schematic illustrating assortative, core-periphery and disassortative community structure

```python
def convert_nx_to_gt(G, partition):
    """
        convert a networkx network to a graph-tool network
        Parameters: networkx network G
            partition: list
        Return: graph-tool network
    """
    N = G.number_of_nodes()
    
    ## create a graph
    g = gt.Graph(directed=nx.is_directed(G))
    ## define node properties
    ## name: id of user
    par = g.vp["partition"] = g.new_vp("int")
        
    node_add = defaultdict(lambda: g.add_vertex())
    
    ## add all nodes first
    for i_d in range(N):
        user_name = i_d
        n=node_add[user_name]
        par[n] = partition[i_d]
        
    ## add all users as nodes
    ## add all retweet/reply as links 
    for i_d in range(N):
        user_name = i_d
        neighbors = G.neighbors(i_d)

        n=node_add[user_name]
        for nei in neighbors:
            n2=node_add[nei]
            g.add_edge(n,n2)

    return g;
```

```python
ggt = convert_nx_to_gt(g,[0]*n+[1]*n)
state = gt.BlockState(ggt,ggt.vp["partition"])
block_g = state.get_block_state().g
col = block_g.vp["color"] = block_g.new_vp("string")
nr = block_g.vp["nr"] = block_g.new_vp("int")
vlist = [v for v in block_g.vertices()]
col[vlist[0]] = '#E75959'
col[vlist[1]] = '#5A5AE7'
for v in reversed(sorted(vlist)):
    nr[v] = state.get_nr()[v]
ers = block_g.ep["ers"] = block_g.new_ep("float")
for e in block_g.edges():
    s = e.source()
    t = e.target()
    if s != t:
        ers[e] = state.get_ers()[e]/nr[s]/nr[t]*10
    else:
            ers[e] = state.get_ers()[e]/nr[s]/(nr[s]-1)*2*10
    print(ers[e])
```

# Case Study

```python
def create_vp_name(g,vp_name):
    title = g.vp[vp_name] = g.new_vp("string")
    i = 0
    for v in g.vertices():
        title[v] = str(i)
        i += 1
        
    return;

def preprocess(G,filename,vp_name = None,ep_name = None):
    create_vp_name(G,"title")
    G = gt.extract_largest_component(G, directed = False, prune = True)
    gt.remove_self_loops(G)
    G.save(filename)
    return G;
```

```python
def create_par_vp(G,par_name,par_list):
    """
        create graph-tool vertex property which stores partition informaiton
        Parameters:
            G: graph-tool network
            par_name: string name for the vertex property
            par_list: partition of node
        Return: partition vertex property
    """
    par_vp = G.vp[par_name] = G.new_vp("int")
    vlist = [v for v in G.vertices()]
    for i in range(len(vlist)):
        par_vp[vlist[i]] = par_list[i]
    return par_vp;
```
